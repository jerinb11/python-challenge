```python
import os
import csv
```

```python
import pandas as pd
```

```python
budget_path = os.path.join("Resources", "budget_data.csv")
print(budget_path)
```

```python
pybank_data_df = pd.read_csv(budget_path)
pybank_data_df
```

```python
months_df = pybank_data_df.loc[:, ["Date"]]
months_df
```

```python
months_df.count()
```

```python
#TOTAL MONTHS
total_months = len(months_df)
```

```python
total_df = pybank_data_df.loc[:, ["Profit/Losses"]]
total_df
```

```python
#storing a variable won't output but it'll store the variable 
#TOTAL PROFIT
total_amount = total_df["Profit/Losses"].sum()
```

```python
#Set up for calculating greatest profit & greatest loss
changes = [0]
prev_row = 0
for i,row in pybank_data_df.iterrows():
    if i==0:
        prev_row = int(row["Profit/Losses"])
    else:
        diff = int(row["Profit/Losses"])-prev_row
        changes.append(diff)
        prev_row = int(row["Profit/Losses"])
len(changes)
```

```python
#AVERAGE CHANGE
average_change=sum(changes)/len(changes)
```

```python
#create a new column for the 'difference'
pybank_data_df["difference"] = changes
```

```python
pybank_data_df
```

```python
#Sorting from highest to lowest 'difference'
average_changes = pybank_data_df.reset_index().sort_values("difference",ascending=False)
average_changes
```

```python
#GREATEST INCREASE
greatest_increase=average_changes.iloc[[0,85]]['difference'].values[0]
#INCREASE MONTH
month_increase=average_changes.iloc[[0,85]]['Date'].values[0]
#GREATEST DECREASE
greatest_decrease=average_changes.iloc[[0,85]]['difference'].values[1]
#DECREASE MONTH
month_decrease=average_changes.iloc[[0,85]]['Date'].values[1]
```

```python
with open ("budget.txt", "w") as file:
    file.write(f"\nFinancial Analysis\n")
    file.write(f"----------------------------\n")
    file.write(f"Total Months: {total_months}\n")
    file.write(f"Total: ${total_amount}\n")         
    file.write(f"Average  Change: ${average_change:.2f}\n")
    file.write(f"Greatest Increase in Profits: {month_increase} (${greatest_increase})\n")         
    file.write(f"Greatest Decrease in Profits: {month_decrease} (${greatest_decrease})\n")          
```
